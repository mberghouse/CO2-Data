```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
infile1 = "../../raw_data/Calhoun/R1C1/cczo_hydroprobe_data.xlsx"
infile2 = "../../raw_data/Calhoun/R1C1/CCZO_soil_gas_data.xlsx" 


df1 = pd.read_excel(infile1, na_values='NAN',  parse_dates=[0], index_col=1) 
df2 = pd.read_excel(infile2, na_values='NAN',  parse_dates=[0], index_col=0) 

```

```python

df1=df1.loc[df1['Plot'] == 'R1C1'] 
df1
```

```python
df2=df2.loc[df2['Site'] == 'R1'] 
df2=df2.loc[df2['Plot'] == 'C1'] 
```

```python
df2=df2.drop(columns='Comments')
df2=df2.drop(columns='Time')

df2=df2.drop(columns='Site')
df2=df2.drop(columns='Plot')
df2
```
